# Equations of Motion of Biped with Arms


```python
# %reset
# DERIVE THE EOMs SYMBOLICALLY ----------------------------------------------------------------------------------------------

# import libraries
import sympy as sym
import numpy as np

sym.init_printing()
from IPython.display import display #for pretty printing

# create symbolic variables
# b - body, a - arm, p - pelvis, l - leg
# _l - left, _r - right

# system parameters
g = sym.symbols('g')
mb,ma_l,ma_r,mp,ml1_l,ml2_l,ml1_r,ml2_r = sym.symbols(['m_{body}','m_{arm_{left}}','m_{arm_{right}}','m_{pelvis}',
                                                       'm_{leg1_{left}}','m_{leg2_{left}}','m_{leg1_{right}}',
                                                       'm_{leg2_{right}}']) # mass

lb,la_l,la_r,lp,ll1_l,ll2_l,ll1_r,ll2_r,l_p2s = sym.symbols(['l_{body}','l_{arm_{left}}','l_{arm_{right}}','l_{pelvis}',
                                                       'l_{leg1_{left}}','l_{leg2_{left}}','l_{leg1_{right}}',
                                                       'l_{leg2_{right}}','l_{p2s}']) # length

Inb,Ina_l,Ina_r,Inp,Inl1_l,Inl2_l,Inl1_r,Inl2_r = sym.symbols(['I_{body}','I_{arm_{left}}','I_{arm_{right}}','I_{pelvis}',
                                                               'I_{leg1_{left}}','I_{leg2_{left}}','I_{leg1_{right}}',
                                                               'I_{leg2_{right}}']) # moment of intertia

# generalized coordinates
xp,yp,thp,thb,tha_l,tha_r,thl_l,thl_r,r_l,r_r = sym.symbols(['x_{pelvis}','y_{pelvis}','\\theta_{pelvis}','\\theta_{body}',
                                                             '\\theta_{arm_{left}}','\\theta_{arm_{right}}',
                                                             '\\theta_{leg_{left}}','\\theta_{leg_{right}}',
                                                             'r_{left}','r_{right}']) 

dxp,dyp,dthp,dthb,dtha_l,dtha_r,dthl_l,dthl_r,dr_l,dr_r = sym.symbols(['\dot{x}_{pelvis}','\dot{y}_{pelvis}','\dot{\\theta}_{pelvis}','\dot{\\theta}_{body}',
                                                                       '\dot{\\theta}_{arm_{left}}','\dot{\\theta}_{arm_{right}}',
                                                                       '\dot{\\theta}_{leg_{left}}','\dot{\\theta}_{leg_{right}}',
                                                                       '\dot{r}_{left}','\dot{r}_{right}']) 

ddxp,ddyp,ddthp,ddthb,ddtha_l,ddtha_r,ddthl_l,ddthl_r,ddr_l,ddr_r = sym.symbols(['\ddot{x}_{pelvis}','\ddot{y}_{pelvis}','\ddot{\\theta}_{pelvis}','\ddot{\\theta}_{body}',
                                                                       '\ddot{\\theta}_{arm_{left}}','\ddot{\\theta}_{arm_{right}}',
                                                                       '\ddot{\\theta}_{leg_{left}}','\ddot{\\theta}_{leg_{right}}',
                                                                       '\ddot{r}_{left}','\ddot{r}_{right}']) 

q = sym.Matrix([[xp],[yp],[thp],[thb],[tha_l],[tha_r],[thl_l],[thl_r],[r_l],[r_r]])
dq = sym.Matrix([[dxp],[dyp],[dthp],[dthb],[dtha_l],[dtha_r],[dthl_l],[dthl_r],[dr_l],[dr_r]])
ddq = sym.Matrix([[ddxp],[ddyp],[ddthp],[ddthb],[ddtha_l],[ddtha_r],[ddthl_l],[ddthl_r],[ddr_l],[ddr_r]])

# forces
# total joint action = actuator + rebound, but that will be dealt with elsewhere
F_a_l,F_a_r,GRFx_l,GRFy_l,GRFx_r,GRFy_r = sym.symbols(['F_{left}','F_{right}','G_{x_{left}}','G_{y_{left}}','G_{x_{right}}','G_{y_{right}}']) 
tau_b,tau_a_l,tau_a_r,tau_l_l,tau_l_r = sym.symbols(['\\tau_{body}','\\tau_{arm_{left}}','\\tau_{arm_{right}}','\\tau_{leg_{left}}','\\tau_{leg_{right}}'])

# STEP 1: position vectors ri = [x,y,theta] (world frame)
rp = sym.Matrix([[xp],
                 [yp],
                 [thp]])

# left legs
rl1_l = sym.Matrix([[xp - 0.5*lp*sym.cos(thp) + 0.5*ll1_l*sym.sin(thp + thl_l)],
                    [yp - 0.5*lp*sym.sin(thp) - 0.5*ll1_l*sym.cos(thp + thl_l)],
                    [thp + thl_l]])

rl2_l = sym.Matrix([[xp - 0.5*lp*sym.cos(thp) + (0.5*ll1_l+r_l)*sym.sin(thp + thl_l)],
                    [yp - 0.5*lp*sym.sin(thp) - (0.5*ll1_l+r_l)*sym.cos(thp + thl_l)],
                    [thp + thl_l]])

# right legs
rl1_r = sym.Matrix([[xp + 0.5*lp*sym.cos(thp) + 0.5*ll1_r*sym.sin(thp + thl_r)],
                    [yp + 0.5*lp*sym.sin(thp) - 0.5*ll1_r*sym.cos(thp + thl_r)],
                    [thp + thl_r]])

rl2_r = sym.Matrix([[xp + 0.5*lp*sym.cos(thp) + (0.5*ll1_r+r_r)*sym.sin(thp + thl_r)],
                    [yp + 0.5*lp*sym.sin(thp) - (0.5*ll1_r+r_r)*sym.cos(thp + thl_r)],
                    [thp + thl_r]])

# body
rb = sym.Matrix([[xp + 0.5*lb*sym.sin(thp+thb)],
                 [yp + 0.5*lb*sym.cos(thp+thb)],
                 [thp + thb]])

# left arm - 2.5/4 heads from pelvils to shoulder (total is pelvis to top of head)
ra_l = sym.Matrix([[xp + l_p2s*sym.cos(thb+thp) - 0.5*la_l*sym.sin(thb+thp+tha_l)],
                   [yp + l_p2s*sym.sin(thb+thp) + 0.5*la_l*sym.cos(thb+thp+tha_l)],
                   [thp + thb + tha_l]])

# right arm
ra_r = sym.Matrix([[xp + l_p2s*sym.cos(thb+thp) + 0.5*la_r*sym.sin(thb+thp+tha_r)],
                   [yp + l_p2s*sym.sin(thb+thp) - 0.5*la_r*sym.cos(thb+thp+tha_r)],
                   [thp + thb + tha_r]])

# the Jacobians
Jp    = rp.jacobian(q)
Jl1_l = rl1_l.jacobian(q)
Jl2_l = rl2_l.jacobian(q)
Jl1_r = rl1_r.jacobian(q)
Jl2_r = rl2_r.jacobian(q)
Jb    = rb.jacobian(q)
Ja_l  = ra_l.jacobian(q)
Ja_r  = ra_r.jacobian(q)

# STEP 2: generate expressions for the system space velocities from the jacobians
vp    = Jp*dq 
vl1_l = Jl1_l*dq
vl2_l = Jl2_l*dq
vl1_r = Jl1_r*dq
vl2_r = Jl2_r*dq
vb    = Jb*dq
va_l  = Ja_l*dq
va_r  = Ja_r*dq

# STEP 3: generate expressions for the kinetic and potential energy
# mass vectors
Mp    = sym.Matrix([[mp,mp,Inp]])
Ml1_l = sym.Matrix([[ml1_l,ml1_l,Inl1_l]])
Ml2_l = sym.Matrix([[ml2_l,ml2_l,Inl2_l]])
Ml1_r = sym.Matrix([[ml1_r,ml1_r,Inl1_r]])
Ml2_r = sym.Matrix([[ml2_r,ml2_r,Inl2_r]])
Mb    = sym.Matrix([[mb,mb,Inb]])
Ma_l  = sym.Matrix([[ma_l,ma_l,Ina_l]])
Ma_r  = sym.Matrix([[ma_r,ma_r,Ina_r]])

T = 0.5*Mp*sym.matrix_multiply_elementwise(vp,vp) + 0.5*Mb*sym.matrix_multiply_elementwise(vb,vb) +\
    0.5*Ml1_l*sym.matrix_multiply_elementwise(vl1_l,vl1_l) + 0.5*Ml2_l*sym.matrix_multiply_elementwise(vl2_l,vl2_l) + \
    0.5*Ml1_r*sym.matrix_multiply_elementwise(vl1_r,vl1_r) + 0.5*Ml2_r*sym.matrix_multiply_elementwise(vl2_r,vl2_r) + \
    0.5*Ma_l*sym.matrix_multiply_elementwise(va_l,va_l) + 0.5*Ma_r*sym.matrix_multiply_elementwise(va_r,va_r)

T = T[0].simplify()
V = mp*g*rp[1] + mb*g*rb[1] + ml1_l*g*rl1_l[1] + ml2_l*g*rl2_l[1] + ml1_r*g*rl1_r[1] + ml2_r*g*rl2_r[1]+ ma_l*g*ra_l[1] + ma_r*g*ra_r[1]
V = V.simplify()

# STEP 4: calculate each term of the Lagrange equation
# term 1
Lg1 = sym.zeros(len(q),1)
for i in range(len(q)):
    dT_ddq = sym.Matrix([sym.diff(T,dq[i])]) # get partial of T in dq_i
    Lg1[i] = dT_ddq.jacobian(q)*dq + dT_ddq.jacobian(dq)*ddq #...then get time derivative of that partial

# term 3
Lg3 = sym.Matrix([T]).jacobian(q).transpose() # partial of T in q

# term 4
Lg4 = sym.Matrix([V]).jacobian(q).transpose() # partial of U in q

# # STEP 5: generalized forces

# # force vectors for each link
Tau_a_l = sym.Matrix([[0],[0],[tau_a_l-tau_b]])
Tau_a_r = sym.Matrix([[0],[0],[tau_a_r-tau_b]])
Tau_b   = sym.Matrix([[0],[0],[tau_b-(tau_a_l+tau_a_r)]])
Tau_p   = sym.Matrix([[0],[0],[-(tau_l_l+tau_l_r)-(tau_b)]])
Tau_l_l = sym.Matrix([[0],[0],[tau_l_l]])
Tau_l_r = sym.Matrix([[0],[0],[tau_l_r]])

F_l1_l = sym.Matrix([[-F_a_l*sym.sin(thp+thl_l)],[ F_a_l*sym.cos(thp+thl_l)],[0]])
F_l2_l = sym.Matrix([[ F_a_l*sym.sin(thp+thl_l)],[-F_a_l*sym.cos(thp+thl_l)],[0]])

F_l1_r = sym.Matrix([[-F_a_r*sym.sin(thp+thl_r)],[ F_a_r*sym.cos(thp+thl_r)],[0]])
F_l2_r = sym.Matrix([[ F_a_r*sym.sin(thp+thl_r)],[-F_a_r*sym.cos(thp+thl_r)],[0]])

GRF_l2_l = sym.Matrix([[GRFx_l],[GRFy_l],[0.5*ll2_l*GRFx_l*sym.cos(thp+thl_l)+0.5*ll2_l*GRFy_l*sym.sin(thp+thl_l)]])
GRF_l2_r = sym.Matrix([[GRFx_r],[GRFy_r],[0.5*ll2_r*GRFx_r*sym.cos(thp+thl_r)+0.5*ll2_r*GRFy_r*sym.sin(thp+thl_r)]])

# pelvis and body, arms, left legs, right legs. 
Q = sym.zeros(len(q),1)
for j in range(len(q)):
    Q[j] =  Tau_p.transpose()*Jp[:,j]+Tau_b.transpose()*Jb[:,j] + \
            Tau_a_l.transpose()*Ja_l[:,j]+Tau_a_r.transpose()*Ja_r[:,j] + \
            (Tau_l_l+F_l1_l).transpose()*Jl1_l[:,j]+(F_l2_l+GRF_l2_l).transpose()*Jl2_l[:,j] + \
            (Tau_l_r+F_l1_r).transpose()*Jl1_r[:,j]+(F_l2_r+GRF_l2_r).transpose()*Jl2_r[:,j]     

for j in range(len(q)):
    Q[j] = Q[j].simplify()
    
# AND combine!
EOM = Lg1 - Lg3 + Lg4 - Q

EOMs = sym.zeros(len(q),1)
for j in range(len(q)):
    EOMs[j] = EOM[j].simplify()
```

```python
# Auxiliary Variables - foot position and velocity in board and global frame

# system variables
xm,ym,thm,thl,r = sym.symbols(['x_{mono}','y_{mono}','\\theta_{mono}','\\theta_{link1}','r_{mono}']) # monopod in global frame
dxm,dym,dthm,dthl,dr = sym.symbols(['\dot{x}_{mono}','\dot{y}_{mono}','\dot{\\theta}_{mono}','\dot{\\theta}_{link1}','\dot{r}_{mono}'])
lb,l1,l2 = sym.symbols(['l_{board}','l_{link1}','l_{link2}']) # lenghts of links
xb,yb,thb = sym.symbols(['x_{board}','y_{board}','\\theta_{board}']) # board in global frame
dxb,dyb,dthb = sym.symbols(['\dot{x}_{board}','\dot{y}_{board}','\dot{\\theta}_{board}'])

q = sym.Matrix([[xb],[yb],[thb],[xm],[ym],[thm],[thl],[r]])
dq = sym.Matrix([[dxb],[dyb],[dthb],[dxm],[dym],[dthm],[dthl],[dr]])

thA = thm + thl
L = 0.5*l1+0.5*l2+r

pb = sym.Matrix([[xb], [yb]])
pfl = sym.Matrix([[xm - 0.5*lb*sym.cos(thm) + L*sym.sin(thA)], [ym - 0.5*lb*sym.sin(thm) - L*sym.cos(thA)]]) # position of left foot in global frame
pfr = sym.Matrix([[xm + 0.5*lb*sym.cos(thm) + L*sym.sin(thA)], [ym + 0.5*lb*sym.sin(thm) - L*sym.cos(thA)]]) # position of right foot in global frame

Rz_thb = sym.Matrix([[sym.cos(thb), sym.sin(thb)], # z-rotation by thA
                     [-sym.sin(thb), sym.cos(thb)]])

pfl_b = Rz_thb*(pfl-pb) # position of left foot in board frame
pfr_b = Rz_thb*(pfr-pb) # position of right foot in board frame

# Velocity of left foot in board frame
vfl_b = pfl_b.jacobian(q)*dq
vfl_bx = vfl_b[0].simplify()
# Velocity of right foot in board frame
vfr_b = pfr_b.jacobian(q)*dq
vfr_bx = vfr_b[0].simplify()
```

```python
# Saving equations of motion so you don't have to generate them again
import pickle as pkl

data = {"EOMs":EOMs, "pfoot_left":pfl, "pfoot_left_board":pfl_b, "vfootx_left_board":vfl_bx, "pfoot_right":pfr, "pfoot_right_board":pfr_b, "vfootx_right_board":vfr_bx}

outfile = open('biped_with_arms_EOMs','wb')
pkl.dump(data,outfile)
outfile.close()
```
